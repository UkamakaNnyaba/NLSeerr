```python
import pandas as pd
```

```python
df = pd.read_csv('../data/yamagishi_kaneko_NLS_table.csv')
```

```python
df.head(9)
```

```python
# Remove redundant rows, columns and rename some columns 

df_cleaned = df.iloc[3:,:]
df_cleaned.reset_index(drop=True, inplace=True)
df_cleaned.columns = df_cleaned.iloc[0]
df_cleaned = df_cleaned.drop(df_cleaned.index[0])
df_cleaned = df_cleaned.drop(['Name', 'Evidence', 'ECO code'],axis=1)
df_cleaned = df_cleaned.drop(df_cleaned.columns[[0,-1]],axis=1)
df_cleaned = df_cleaned.rename(columns={'UniProt ID':'ID'})
print(df_cleaned)
```

```python
print(df_cleaned.columns)
```

```python
# Full sequences downloaded from UniProt
file_path='../data/UniProt_data.tsv'

df1 = pd.read_csv(file_path, sep='\t')

df1 = df1.drop(['Entry','Length'], axis = 1)
df1 = df1.rename(columns={'From':'ID'})
df1.head(5)
```

```python
# Append full protein sequences to the cleaned yamagishi_kaneko dataset
merged_df = pd.merge(df_cleaned, df1, on='ID', how='outer')
print(merged_df.head())
```

```python
merged_df.to_csv('../data/data_NLS.csv', index=False)
```
