```python
from Bio import SeqIO
import csv
```

```python
#Create function that converts FASTA file to CSV

def fasta_to_csv(fasta_file, csv_file):
    with open(fasta_file, "r") as fasta_handle, open(csv_file, "w", newline="") as csv_handle:
        writer = csv.writer(csv_handle)
        writer.writerow(["ID", "Sequence"])
        for record in SeqIO.parse(fasta_handle, "fasta"):
            writer.writerow([record.id, str(record.seq)])

# Usage
fasta_to_csv("idmapping_2024_04_16.fasta", "mitochondrial_dataset.csv")
```
